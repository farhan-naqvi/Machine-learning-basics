# Heatmap

```python
# let us find out what is the correlation between the data

df.corr()# in tabular format 
```

```python
#in visual format, we can use heatmap
plt.figure(figsize=(12,12))
sns.heatmap(df.corr())
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(df.corr(),linewidths=1,annot=True)
```

# Joint plot

```python
#plt.figure(figsize=(12,12))
#sns.color_palette("flare", as_cmap=True)
#sns.color_palette("flare")
sns.jointplot(x='carat',y='price',data=df,height=10,color="g")
```

```python
sns.jointplot(x='carat',y='price',data=df,kind="hex",height=10,color="r")
```

```python
sns.jointplot(x='carat',y='price',data=df,kind="reg",height=10,color="g")
```

```python
sns.jointplot(x='carat',y='price',data=df,kind="resid",height=15,color="g")
```
